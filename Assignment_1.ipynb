```python
import pandas as pd

# Creating a sample dataset
data = {
    "Name": ["Jean", "Jehue", "Nana"],
    "Age": [47, 30, 32],
    "Score": [85, 90, 88]
}

df = pd.DataFrame(data)

print (df)
```

# Dataset Description

This dataset contains information about three individuals:
- Name: The person's name
- Age: Their age
- Score: Their performance score

The data is simple and useful for demonstrating basic Pandas operations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot
sns.barplot(x="Name", y="Score", data=df)

plt.title("Scores of Individuals")
plt.xlabel("Name")
plt.ylabel("Score")

plt.show()
```

## Visualization Explanation

The bar chart shows the scores of each individual.
- Jehue has the highest score.
- Jean and Nana have slightly lower scores.

This visualization helps quickly compare values across categories.
